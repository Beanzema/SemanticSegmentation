# MinMax Scaling for Better Performance

```python
# Let's take a look at the dataset
import mmcv
import matplotlib.pyplot as plt

img = mmcv.imread('dataset/satellite_512_images/LC_SN_36602_001.tif')
plt.figure(figsize=(8, 6))
plt.imshow(mmcv.bgr2rgb(img))
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()

single_patch_img = img

single_patch_img = scaler.fit_transform(single_patch_img.reshape(-1, single_patch_img.shape[-1])).reshape(single_patch_img.shape)
```

```python
single_patch_img
```

```python
myArray = (single_patch_img*255).astype(int)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title('Before(Raw Image)')
plt.imshow(img[:,:,:])
plt.subplot(122)
plt.title('After(MinMax Scaler)')
plt.imshow(myArray)
plt.show()
```

```python
import cv2
cv2.imwrite('test.png', myArray)
```
